```
import glob
import re
import itertools




#Klasse zum Speichern der SN und deren zugehörigen Dateiname
class SignLev:

  
  #Konstruktor der Klasse
  #Ihm werden die Liste der SN und der Dateiname übergeben 
    def __init__(self, name, sign_levels):
      
      #Festlegung des Namens 
        self.Name = name
        
        #Festlegung  der SN 
        self.SignLevelsSource = sign_levels
        
        #Initialisierung der SN-Interpreter auf None
        self.SingLevs = None
        
        #Liste von all den Zahlen der SN, die gefunden wurden 
        self.allNumbers = list()
        
        #Liste von all den Zahlen der SN und deren Wiederholungen
        self.NumbersAndOccurences = {}
        
        #Checkt ob die SN eine Zahl ist, wenn ja dann ist etwas bestimmt schiefgelaufen 
        #oder ob die Anzahl der SN gleich null ist
        #In solchen Fällen werden die SN-Intepreter und die maximale bzw. minimale gefundene Zahl auf None gesetzt 
        if type(sign_levels) is int or len(sign_levels) == 0:
            self.SingLevs = None
            self.Max = None
            self.Min = None
        
        #wird eingetreten, wenn alles gut gelaufen ist und die Anzahl der gefundenen SN größer als null ist 
        else:
          
          #Aufruf der Methode zur Festlegung der SN-Interpreter
            self.interpretSignLev()
            
            #Versuch, aus den Zahlen, die maximale und minimale zu bestimmen
            try:
                self.Max = max(self.allNumbers)
                self.Min = min(self.allNumbers)
            
            #Bei einem Fehlschlag wird die maximale bzw. minimale Zahl auf None gesetzt 
            except TypeError as err:
                self.Max = None
                self.Min = None    
            #Sortiert und gruppiert letzendlich die Zahlen nach Einzigartigkeit
            finally:
                self.SortAndgroupByNumber()

                
    #Überschreibung der str-Methode der Klasse             
    def __str__(self):
      
      #Formulierung der anzuzeigenden Nachricht bei Aufruf der str Methode 
        message = ""
        if self.SingLevs == None:
            message = "hat keine\n"
        else:
            for elt in self.SingLevs:
                message = message + str(elt)

        message3 = ""
        message2 = "\nMin = " + str(self.Min) + "\nMax =  " + str(self.Max)
        
        #Formulierung der anzuzeigenden Nachricht für die Zahlen und deren Wiederholungen 
        for key, value in self.NumbersAndOccurences.items():
            message3 += "\nNumber " + str(key) + " has " + str(value) + " occurences\n"

        #Rückgabe der vollständigen Nachricht oder String
        return "\n\n********  Datei : " + str(self.Name) + "\n" + "\n\nSignifikanzniveaus : \n\n" + message + message2 + message3 + "\n**********\n\n"

      
    #Methode zur Erzeugung bzw. Festlegung der SN-Interpreter
    def interpretSignLev(self):
      
      #leere Liste der SN-Interpreter
        self.SingLevs = list()
        
        #Durlauf der Liste der SN
        for signL in self.SignLevelsSource:
          
            #Erzeugung des aktuellen SN-Interpreters
            #Übergabe in dem Konstruktor des SN zum Interpretieren 
            signLInter = SignLevInterpreter(signL)
            
            #Hinzufügen der Zahl des SN-Interpreters in der Liste der gesamten Zahlen der SN
            self.allNumbers.append(signLInter.Number)
            
            #Hinzufügen des aktuellen SN-Interpreters zu der Liste der SN-Interpreter
            self.SingLevs.append(signLInter)

    #Liste zur Sortierung und Gruppierung der gesmaten Zahlen der SN
    #Anschliesßend wird für jede Zahl die Anzahl ihrer Wiederholungen bestimmt
    def SortAndgroupByNumber(self):
        for key, group in itertools.groupby(sorted(self.allNumbers)):
            self.NumbersAndOccurences[key] = len(list(group))
            
class SignLevInterpreter:
   
   #Übergabe des Signifikanzniveaus 
    def __init__ (self, signLev):

        #Trennt das SN mit dem "=" Zeichen
        elts = signLev.split('=')
        #Wenn etwas gefunden wurde dann setzt er den Status auf 0, also signalisiert die gleich
        if len(elts) > 1:
            self.State = 0

        #Wenn nicht dann weiter hier rein    
        else:
            #Trennt das SN mit dem "<" Zeichen
            elts = signLev.split('<')
            #Wenn etwas gefunden wurde dann setzt er den Status auf -1, also signalisiert die  kleiner gleich
            if len(elts) > 1:
                self.State = -1

            
            #Wenn nicht dann weiter hier rein
            else:
                #Trennt das SN mit dem ">" Zeichen
                elts = signLev.split('>')
                #Wenn etwas gefunden wurde dann setzt er den Status auf 1, also signalisiert die  größer gleich
                if len(elts) > 1:
                    self.State = 1
                else:
                    #Wenn nicht dann ist etwas Üngültiges
                    self.State = -2

        #Wenn kein ungültiger Status festgestellt wurde, dann wird die Zahl extrahiert            
        if not self.State == -2:
            subElt = elts[1].strip().replace(' ', '.')
            self.Number = float(subElt)
        
        #Wenn wir eine ungültige Zahl haben, dann setzen wir die Zahl auf None
        else:
            self.Number = None

    
    #Überschreibung der str-Methode
    def __str__ (self):

        message = ""

        #Sukzessiver Aufbau der Ausgabenachricht  je nachdem welcher Status hierbei gilt
        if self.State == 0:
            message = " ist Nullhypothese "
            if self.Number >= 0.05:
                message += " und statistisch nicht signifikant "
            elif self.Number >= 0.01:
                message += " aber statistisch signifikant "
            else:
                message += " aber statistisch hochsignifikant "
        elif self.State == -1:
            if self.Number <= 0.05 and self.Number >= 0.01:
                message = " ist statistisch signifikant "
            else:
                message += " ist statistisch hochsignifikant "
        elif self.State == 1:
            if self.Number < 0.05:
                message += " ist statistisch nicht interpretierbar "
            else:
                message += " ist statistisch nicht signifikant "
        else:
            message += " ist statistisch nicht gültig "

        subMessage = (" = " if self.State == 0 else (
            " > " if self.State == 1 else (" < " if self.State == -1 else "unavailable "))) + str(
            self.Number)
        return "\nDas Testergebnis für das Signifikanzniveau " + str(
            self.Number) + message + ": ( p" + subMessage + " )"

      
#Methode zum Extrahieren der Signifikanzniveaus nach den mit "file_names" übergebenen Dateien      
def get_files_significant_levels(file_names):
  
  #Erzeugung einer leeren Liste für die gefundenen SN nach Dtaieen 
    elts = list()
    
    #Initialisierung der Anzahl der gefundenen Dateien mit keiner Angabe von SN auf null
    counter = 0
    
    
    #Durchlauf der Liste der übergebenen Dateien 
    for fileName in file_names:
      
      #Versuht, die aktuelle Durchlaufdatei mit dem Namen in "fileName" enthalten zu öffnen und daraus die SN zu extrahieren
        try:
          
          #Öffnung der aktuellen Durchlaufdatei mit dem Namen in "fileName" 
            with open(fileName, encoding="utf8", mode='r') as f:
              
              #Auslesen des Inhalts der Datei 
                text = f.read()
                
                #Extrahierung der SN aus der ausgelesenen Datei
                signLev = re.findall(r'p\s*[<,=,>]\s*[0, ][., ]\d+', text)
                
                #Inkrementierung der counter-Variable, Wenn die Anzahl der gefundenen SN von der aktuellen Datei gleich null ist
                if (len(signLev) == 0):
                    counter = counter + 1
                    
                   
                  
                #Erzeugung eines Objetes der Klasse SignLev
                #Im Konstrutor nimmt das Objekt die SN (signLev) und den Dateinamen, zu dem sie hingehören 
                newSingLev = SignLev(fileName, signLev)
                
                #Hinzufügen des erzeugten Objektes mit den SN in der Liste der SN 
                elts.append(newSingLev)
                
        #Wenn etwas schiefgelaufen ist, z.B. Öffnen der Datei gescheitert oder Auslesen des Inhalts der Datei nicht möglich, etc...       
        except:
            
            
            #Erzeugung eines Objetes der Klasse SignLev
            #Im Konstrutor nimmt das Objekt die Zahl -1 und den Dateinamen, zu dem die Zahl hingehört 
            #die Zahl -1 signalisiert,dass etwas bei der Extrahierung schiefgelaufen ist 
            newSingLev = SignLev(fileName, -1)
            
           #Hinzufügen des erzeugten Objektes 
            elts.append(newSingLev)

    return elts, counter


  
  
  #Hauptfunktion, die als erstes aufgerufen wird und von der die anderen Methoden aufzurufen und aufgerufen sind 
def main():
  #Alle Textdateien von dem angegebenen Pfad auflisten ( also deren Pfade)
    files = glob.glob("drive/My Drive/Bachelorarbeit josy/Datensätze/*.txt")
    
    #Anzeige der Anzahl der gefundenen Dateien
    print(str(len(files)) + " Dateien gefunden\n")
    
    #Übergabe der Pfade der gefundenen Textdateien
    #Rückgabe:
    #results: Liste der Signifikanzniveaus entsprechend der Dateien
    #counter: Anzahl der Dateien mit keiner Angabe der Siginifikanzniveaus
    results, counter = get_files_significant_levels(files)
    
    #Anzeige der Anzahl der Dateien mit keiner Angabe der Siginifikanzniveaus
    print("\n" + str(counter) + " Artikel haben kein Signifikanzniveau\n")
    
    #Sukzessive Anzeige der gefundenen Signifikanzniveaus nach Datei bzw. Artikel
    for result in results:
        print(result)
        
#Aufruf der Hauptmethode
main()
```
